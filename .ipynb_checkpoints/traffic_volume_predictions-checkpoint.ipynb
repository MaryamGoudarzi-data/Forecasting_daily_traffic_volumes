```python
import pandas as pd

# Loading traffic data  
traffic = pd.read_csv("Daily_data.csv")

# Checking the column names
print(traffic.columns)

# Checking column data types
print(traffic.dtypes)

# Number of rows
print("Total rows:", len(traffic))

# checking for missing values in each column
print(traffic.isnull().sum())
```

```python
# Load weather data from Open-Meteo API
weather = pd.read_csv("Weather_data.csv")

# Renaming columns in weather dataset
weather.columns = ['date', 'temperature(°C)', 'cloud_cover(%)', 'rain(mm)', 'wind(km/h)']

# Converting date to Datetime
weather['date'] = pd.to_datetime(weather['date'], dayfirst= True)

# Setting the index to date to merge weather data with traffic data using date
weather.set_index('date', inplace=True)

print(weather.head())
```

```python
import pandas as pd
import numpy as np

# Converting date to datetime format
traffic['date'] = pd.to_datetime(traffic['date'])

# Set the date column as the index
traffic.set_index('date', inplace=True)

# Define hourly columns
hour_cols = [f"hour_{str(i).zfill(2)}" for i in range(24)]

# Interpolate missing hourly values before summing them
traffic[hour_cols] = traffic[hour_cols].interpolate(axis=1, limit_direction='both')

# Creating 'traffic_volume' by summing hourly columns
traffic['traffic_volume'] = traffic[hour_cols].sum(axis=1)

# Dropping rows where entire day is still NaN 
traffic = traffic.dropna(subset=['traffic_volume'])

# Adding day of week and month
traffic['day_of_week'] = traffic.index.dayofweek
traffic['month'] = traffic.index.month

# Cyclical encoding for day of week
traffic['dow_sin'] = np.sin(2 * np.pi * traffic['day_of_week'] / 7)
traffic['dow_cos'] = np.cos(2 * np.pi * traffic['day_of_week'] / 7)

# Renaming holiday columns
traffic.rename(columns={'public_holiday': 'is_public_holiday', 'school_holiday': 'is_school_holiday'}, inplace=True)

# Preview of data
print(traffic[['traffic_volume', 'is_public_holiday', 'is_school_holiday', 'day_of_week', 'month']].head())
```

```python
# EDA 1: Traffic volume by hour of the day
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Stacking all 24 hourly columns into a single column 
hour_cols = [f'hour_{str(i).zfill(2)}' for i in range(24)]
hourly_long = traffic[hour_cols].copy()
hourly_long['date'] = traffic.index

# Transforming data from wide to long for boxplot
hourly_long = hourly_long.melt(id_vars='date',
                               value_vars=hour_cols,
                               var_name='hour',
                               value_name='traffic_volume')

# Convert hour string to integer as we want the hour to be a number from 0 to 23, not a string like 'hour_00'
hourly_long['hour'] = hourly_long['hour'].str.extract(r'hour_(\d+)').astype(int)

# Boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(x='hour', y='traffic_volume', data=hourly_long,
            palette='Spectral', showfliers=False)
plt.title('Traffic Volume by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Traffic Volume')
plt.grid(True)
plt.tight_layout()
plt.savefig('hourly_traffic.png')
plt.show()
```

```python
# EDA 2: Average daily traffic over time
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(traffic.index, traffic['traffic_volume'])
plt.title("Daily Traffic Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Traffic Volume")
plt.grid(True)
plt.tight_layout()
plt.savefig('daily_traffic.png')
plt.show()
```

```python
# EDA 3: Seasonal Trends – Monthly Averages
monthly_avg = traffic.groupby('month')['traffic_volume'].mean()
monthly_avg.plot(kind='bar', title='Average Monthly Traffic Volume', figsize=(10, 5), color=plt.cm.tab20c(range(12)))
plt.ylabel("Traffic Volume")
plt.xlabel("Month")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('monthly_traffic.png')
plt.show()
```

```python
# EDA 4: Weekly Pattern – Day of Week Averages
dow_avg = traffic.groupby('day_of_week')['traffic_volume'].mean()
dow_avg.index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
dow_avg.plot(kind='bar', title='Average Traffic Volume by Day of Week', figsize=(10, 5), color=plt.cm.Paired(range(7)))
plt.ylabel("Traffic Volume")
plt.xlabel("Day of Week")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('week_traffic.png')
plt.show()
```

```python
# EDA 5: Holiday vs Non-Holiday
traffic.groupby('is_public_holiday')['traffic_volume'].mean().plot(kind='bar', 
    title='Average Traffic Volume: Non-Holiday vs Holiday',
    xticks=[0, 1], figsize=(8, 4), color=plt.cm.Set1(range(2)))

plt.ylabel("Traffic Volume")
plt.xlabel("Is Public Holiday (0=No, 1=Yes)")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('holiday_traffic.png')
plt.show()

# This figure illustrates the difference in traffic volume on holidays compared to non-holiday days. Non-holiday days show significantly higher and more variable traffic volumes compared to holidays.
```

```python
# Merge traffic and weather data on date index
traffic = traffic.merge(weather, left_index=True, right_index=True)

# Preview
print(traffic[['traffic_volume', 'temperature(°C)', 'cloud_cover(%)', 'rain(mm)', 'wind(km/h)']].head())
```

```python
# Weather-based EDA
import matplotlib.pyplot as plt
import seaborn as sns

# Defining colors
colors = ['royalblue', 'tomato', 'mediumseagreen', 'darkorange']

# Creating 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot1: Traffic vs Rain
sns.scatterplot(ax=axes[0, 0], x='rain(mm)', y='traffic_volume', data=traffic, color=colors[0])
axes[0, 0].set_title("Traffic vs Rain")
axes[0, 0].set_xlabel("Rain (mm)")
axes[0, 0].set_ylabel("Traffic Volume")

# Plot2: Traffic vs Temperature
sns.scatterplot(ax=axes[0, 1], x='temperature(°C)', y='traffic_volume', data=traffic, color=colors[1])
axes[0, 1].set_title("Traffic vs Temperature")
axes[0, 1].set_xlabel("Temperature (°C)")
axes[0, 1].set_ylabel("Traffic Volume")

# Plot3: Traffic vs Cloud Cover
sns.scatterplot(ax=axes[1, 0], x='cloud_cover(%)', y='traffic_volume', data=traffic, color=colors[2])
axes[1, 0].set_title("Traffic vs Cloud Cover")
axes[1, 0].set_xlabel("Cloud Cover (%)")
axes[1, 0].set_ylabel("Traffic Volume")

# Plot4: Traffic vs Wind Gusts
sns.scatterplot(ax=axes[1, 1], x='wind(km/h)', y='traffic_volume', data=traffic, color=colors[3])
axes[1, 1].set_title("Traffic vs Wind Gusts")
axes[1, 1].set_xlabel("Wind (km/h)")
axes[1, 1].set_ylabel("Traffic Volume")

plt.tight_layout()
plt.savefig('weather_traffic.png')
plt.show()
```

```python
# Seasonal Decomposition 
## Time series method to break traffic data into trend + seasonality + residuals (random fluctuations like events)
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(traffic['traffic_volume'], model='additive', period=7)

fig = decomp.plot()
fig.set_size_inches(14, 12)

plt.tight_layout()
plt.show()

# Performed time series decomposition to analyze residuals (unexplained variations).
# The residual plot shows large deviations from the baseline (zero line), confirming spikes and drops in traffic volume.
# These deviations suggest unusual events (e.g., road closures, crashes, parades) that trend and seasonality cannot explain.
# This supports adding a feature (is_unusual) to flag such days and help the model handle them better.
```

```python
# Quick check
print(traffic.columns)
```

```python
# Time-series forecasting
## Model 1: Random forest regression

# Creating lag-based features
traffic['lag_1'] = traffic['traffic_volume'].shift(1)
traffic['lag_7'] = traffic['traffic_volume'].shift(7)
traffic['rolling_mean_7'] = traffic['traffic_volume'].rolling(window=7).mean()

# Creating 'is_unusual' flag ( spike/dip > 30% from 7-day average )
traffic['is_unusual'] = ((traffic['traffic_volume'] - traffic['rolling_mean_7']).abs() / traffic['rolling_mean_7']) > 0.3

# Defining feature columns
features = ['lag_1', 'lag_7', 'rolling_mean_7', 'day_of_week', 'month', 'is_public_holiday', 'is_school_holiday', 'temperature(°C)', 'cloud_cover(%)', 'rain(mm)', 'wind(km/h)', 'is_unusual']

# Dropping any rows with missing values in features or target value
traffic = traffic.dropna(subset=features + ['traffic_volume'])

# Checking sample of final dataset
print(traffic[['traffic_volume'] + features].head(10))

# Train-test split (no shuffling for time series)
split_index = int(len(traffic) * 0.8)
train = traffic.iloc[:split_index]
test = traffic.iloc[split_index:]

# Preparation of features and target variables
X_train = train[features]
y_train = train['traffic_volume']
X_test = test[features]
y_test = test['traffic_volume']

# Training Random Forest Regressor model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predicting and evaluation
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
relative_mae = mae / y_test.mean()
print("Random Forest MAE:", round(mae, 2))
print("Random Forest RMSE:", round(rmse, 2))
print("Random Forest Relative MAE:", round(relative_mae * 100, 3), "%")

# To further improve model performace, unusual day flag was added to the model, so the model now knows when to be cautious due to the spike/dip indicator..
```

```python
# Visualising Predictions vs. Actual
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(test.index, y_test, label="Actual", alpha=0.7)
plt.plot(test.index, y_pred, label="Predicted", alpha=0.7)
plt.title("Actual vs Predicted Daily Traffic Volume")
plt.xlabel("Date")
plt.ylabel("Traffic Volume")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('random_forest_prediction.png')
plt.show()
```

```python
# Random Forest feature importance plot to show which variables had the most influence on the model’s predictions
importances = model.feature_importances_
plt.figure(figsize=(8, 5))
plt.barh(features, importances)
plt.title("Feature Importance - Random Forest")
plt.xlabel("Importance Score")
plt.grid(True)
plt.savefig('rf_feature_important.png')
plt.show()
```

```python
# Model 2: XGBoost with hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

# Defining the parameter grid to search through.
# These are key hyperparameters that influence model complexity, learning speed, and performance:
# - n_estimators: number of trees (more trees = better fit but longer training)
# - max_depth: depth of each tree (controls how complex each tree can get)
# - learning_rate: how much each tree adjusts predictions (smaller values = slower but more stable)
# - subsample: % of rows used for each tree (helps prevent overfitting)
# - colsample_bytree: % of columns used per tree (another regularization method)
param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

# Initializing the XGBoost model
# XGBoost is a gradient boosting machine learning algorithm that builds trees sequentially to correct errors from previous ones.
xgb = XGBRegressor(random_state=42)

# Running RandomizedSearchCV to find the best hyperparameters
# - n_iter=20: only try 20 random combinations (faster than full grid search)
# - cv=3: 3-fold cross-validation to assess how well each parameter set generalizes
# - scoring: using negative MAE (lower MAE = better)
# - n_jobs=-1: use all CPU cores to speed up training
random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=20,
    cv=3,
    scoring='neg_mean_absolute_error',
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fitting the search to training data
# This will train 20 different XGBoost models using different parameter combinations and evaluate them using cross-validation.
random_search.fit(X_train, y_train)

# Best parameters and score and its corresponding MAE
# This helps select the best-performing model configuration for final training.print("Best Parameters:", random_search.best_params_)
print("Best MAE (CV):", -random_search.best_score_)
```

```python
# Final XGBoost model using the best parameters found from RandomizedSearchCV
xgb_model = XGBRegressor(
    subsample=0.6,
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    colsample_bytree=1.0,
    random_state=42
)

# Fit the model to the training data
xgb_model.fit(X_train, y_train)

# Predict traffic volume on the test set
xgb_preds = xgb_model.predict(X_test)

# Evaluate performance using common metrics
xgb_mae = mean_absolute_error(y_test, xgb_preds) #average difference between the predicted and actual traffic volumes so here our model is off by 1901 vehicles per day
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_preds)) #penalizes larger errors more heavily than MAE
xgb_relative_mae = (xgb_mae / y_test.mean()) * 100 # On average, the model’s predictions are off by 5.5% compared to a typical day’s traffic volume. 
print("Tuned XGBoost MAE:", round(xgb_mae, 2))
print("Tuned XGBoost RMSE:", round(xgb_rmse, 2))
print("XGBoost Relative MAE:", round(xgb_relative_mae, 3), "%")
```

```python
# LightGBM Forecasting
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error

# Use the same features as before
features = ['lag_1', 'lag_7', 'rolling_mean_7', 'day_of_week', 'month',
            'is_public_holiday', 'is_school_holiday',
            'temperature(°C)', 'cloud_cover(%)', 'rain(mm)', 'wind(km/h)', 'is_unusual']

# Dropping missing values
data = traffic.dropna(subset=features + ['traffic_volume'])

# Train-test split
split = int(len(data) * 0.8)
train = data.iloc[:split]
test = data.iloc[split:]

# Preparing inputs
X_train = train[features]
y_train = train['traffic_volume']
X_test = test[features]
y_test = test['traffic_volume']

# Training the model
model = lgb.LGBMRegressor(n_estimators=300, learning_rate=0.05, max_depth=6, random_state=42)
model.fit(X_train, y_train)

# Predicting and evaluation metrics
y_pred = model.predict(X_test)
lgb_mae = mean_absolute_error(y_test, y_pred)
lgb_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
lgb_relative_mae = (mae / y_test.mean()) * 100
print("LightGBM MAE:", round(lgb_mae, 2))
print("LightGBM RMSE:", round(lgb_rmse, 2))
print("LightGBM Relative MAE:", round(lgb_relative_mae, 3), "%")
```

```python
# Compare model performance
import pandas as pd

# Model comparison dataframe for readability
model_comparison = pd.DataFrame({
    'Model': ['Random Forest', 'XGBoost (Tuned)', 'LightGBM'],
    'MAE': [mae, xgb_mae, lgb_mae],
    'RMSE': [rmse, xgb_rmse, lgb_rmse],
    'Relative MAE (%)': [relative_mae * 100, xgb_relative_mae, lgb_relative_mae]
})

# Rounding for readability
model_comparison = model_comparison.round(3)

print("🔎 Model Performance Summary:")
print(model_comparison)

# Saving to CSV 
model_comparison.to_csv("model_performance_summary.csv", index=False)
```
